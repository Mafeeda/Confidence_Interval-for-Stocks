## Q. Find confidence intervals for Beml and Glaxo stocks

```python
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import norm
```

```python
beml = pd.read_csv('beml_df.csv')
beml
```

```python
glaxo = pd.read_csv('glaxo_df.csv')
glaxo
```

```python
# 95% CIE [Confidence Interval estimate] for Beml 

beml_ci = stats.norm.interval(0.95,beml.gain.mean(),beml.gain.std()/(1738**0.5))
print('Beml Gain at 95% CI is:',np.round(beml_ci,6))
```

```python
# 95% CIE [Confidence Interval estimate] for Glaxo

glaxo_ci=stats.norm.interval(0.95,glaxo.gain.mean(),glaxo.gain.std()/(1734**0.5))
print('Glaxo Gain at 95% CI is:', np.round(glaxo_ci,6))
```
